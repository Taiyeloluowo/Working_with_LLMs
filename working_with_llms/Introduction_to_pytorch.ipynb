##**DEEP LEARNING**

**What is Deep Learning?**

Deep learning is a subset of machine learning where algorithms learn from vast amounts of data using ***neural networks***, which are inspired by the ***human brain***.

It is called **deep** becuase the neural networks have many layers, wwhich helps in understanding complex patterns in data.

Neural networks are composed of ***layers of nodes*** (also known as neurons), each layer transforming the data before passing it on to the next layer.

**Neural Networks**

A neural network consists of three types of layers:

1. **Input layer**: Receives the data(image,videos,text, numbers)
2. **Hidden layers**: Intermediate layers that process the data.This is where the learning happens by transforming the inputs. The more the hidden layers, the deeper the networks.
3. **Output layer**: Produces the final prediction.

**How Does Neural Network Learn?**

The network learns by adjsting weights(parameters) and biases as it processes data through these layers.

**How Does the Neurons Work**
* Each neurons performs a mathematical operation;
1. Takes in input, and each input is multiplied by a wieght.
2. It adds up all of the weighted inputs.
3. Applies an activation function, that is, deciding if the neuron should pass a signal forward or now.
4. Then passesthe output to the next layer.

```c
y = f(w1x1 + w2x2 + b)

x1,x2 are the inputs
w1,w2 are the weights(learned during training, they decide how important each input is)
b is the bias(helps to improve flexibility, that is, allows the model to adjust predictions independently of the input)
f is an activation function(it decides the neurons response)

In sum, the goal of the training is to find the best weights and biases that minimize errors.
```

**Activation Function**
These functions decides if a neuron should "activate" and pass information.
1. ReLU (Rectified linear Unit): It is the most common in hidden layers
```c
f(x) = max(0, x)
```
2. Sigmoid: This is good for probability outputs
```c
f(x) = 1/1 + e^-x
```
3. Softmax: This is used in multiclass classification problems.



**Key points about training a neural network:**

**Data preparation:**
The first step is to gather and prepare a large dataset with labeled examples relevant to the task you want the network to learn.

**Network architecture design:**
You need to define the structure of the neural network, including the number of layers, neurons per layer, and activation functions.

**Weight initialization:**
Before training, the network's weights are randomly initialized.

**Forward propagation:**
During training, input data is fed through the network, calculating the output at each layer.

**Loss function:**
The difference between the network's predicted output and the expected output is measured using a loss function, which quantifies the error.

**Backpropagation:**
This is the core learning algorithm where the error signal is propagated backward through the network, allowing the weights to be adjusted proportionally to reduce the loss.

**Gradient descent:**
A common optimization algorithm used to update the weights, where the network iteratively moves in the direction of the steepest descent to minimize the loss function.

**Epochs:**
A single pass through the entire training dataset is called an epoch, and the training process usually involves multiple epochs to improve accuracy.

**Important considerations when training a neural network:**

1. **Hyperparameter tuning:**
Selecting the right values for learning rate, batch size, and other parameters can significantly impact training performance.

2. **Overfitting:**
When the network learns the training data too well, it might not generalize well to unseen data, so techniques like regularization are often used to prevent this.
3. **Validation set:**
A portion of the data is held aside to evaluate the network's performance during training, preventing overfitting and helping to choose the best model.



**Training a Neural Network**

The process involves:
1. **Forward Pass**: The data is passed through the network to get the outputs. That is, the data from moves from input to output through the hidden layers.The network makes prediction based on initial weights and biases.

2. **Loss Calculation**: The output is compared to the true value, amd a  loss is computed.Common loss function is **MSE** for regrssion and **Cross-Entropy Loss** for classification.

3. **Backward Pass(backpropagation)**: The losss is propagated backward through  the network, calculating the gradient(how each weight contributed to the error).
The network learns by adjsuting weights and biases using backpropagation. It uses **Gradient Descent** a mathematical optimization method to minimize loss


4. **Optimzation**: Using an optimizer, the models's parameter(weights and biases) are updated to minimize the loss. examples of optimizers are **Stochastic Gradient Descent(SDG)** and **Adaptive Moment Estimation(Adam)**

5. The process repeats until network learns well enough to make good predictions.

**Optimization Algorithms**

They are like sets of rules that decides how weights and biases are adjusted to minimize error(loss) in a model.
* SDG: One of the simplest and most commonly used optimization methods in deep learning. It updates weights based on the gradient of the loss function with respect to each weight.

**How SDG works**

1. Compute the gradient, that is how much each weight contributes to the error.
2. Updates the weights, that is adjusting it to reduce error
3. It keeps updating until the error is minimiized.
It is simple and works well for small datasets.And it requires careful tuning of the learning rate.

**How Adam works**

This is an advance d optimizer that automatically adjusts the learning rate for each wieght using **momentum** and **adaptive learning rates**

1. Remembers past gradients to smooth updates, that is, moving averages of the gradients or we could called it momentum.
2. Adjusts learning rate individually for each parameter, that is, helps for faster conveergence.
3. comebines the features of  SDG and RMSProp, that is, making it more efficeint.

It works very well for large datasets and deep networks, it converges faster than SDG. and less sensitive to learning rate choice.

#### **...More on Deep Learning**

* The ability of a neural network to ingest data and extract useful representations on the basis of examples is what makes deep learning so powerful.

* Often, these automatically created

**COMPARING DEEP LEARNING TO MACHINE LEARNING**
* With machine learning, we see a practitioner busy defining and engineering features, and feeding them to a learning algorithm; the results on the task will be as good as the features the practitioner engineers.

* With deep learning, the raw data is fed to an algorithm that extracts hierarchical features automatically, guided by the optimization of its own performance on the task; the results will be as good as the ability of the practitioner to drive the algorithm toward its goal.

In sum, we can basically say that deep learning exchanges the need to handcraft features for an increase in data and computational requirements.

## **Automatic Differentiation (autograd)**

One of the most powerful features of pytorch is **autograd**, which is its automatic differentiation system. Autograd allows you to **compute gradients automatically**, which is crucial for **backpropagation** in neural networks.

**gradients(derivatives)**

* These are the values that tell you how much change is needed for each parameter (such as weights and biases) to minimize the loss function.
* A gradient is a vector that contains the partial derivatives of a function with respect to its inputs. When training a neural network, the gradient tells us how to update the model's parameters to improve the prediction.
* If a parameter (like a weight in your network) is too high, the gradient will indicate that you need to decrease it.
* If a parameter is too low, the gradient will tell you to increase it.
* During training, the optimizer uses these gradients to adjust the parameters to minimize the error (loss) in the model.

This means that when you define a model in pytorch, the framework can automatically compute the gradients of all parameters during training- how much each weight in the network should change to minimise the loss

when you create a tensor with `requires_grad=True`, PyTorch starts tracking all operations that are performed on it. This allows it to calculate the gradients later using automatic differentiation.

````python
# lets create a tensor and enable gradients
x = torch.tensor([2.0, 3.0, 4.0, 0.7], requires_grad = True)

# define a function
y = x[0]**2 + x[1]**2

# compute the gradient(derivative)
y.backward()

# lets access the gradient of x
print(x.grad)
````

During the training of neural networks, you need to adjust model parameters (weights) based on how well the model is performing (via backpropagation). Autograd makes this process efficient by automatically calculating gradients.

````python
x = torch.tensor(3.0, requires_grad = True)

# define a function
y = x**2 + 2*x + 1
# compute the gradient(derivative)
y.backward() #  the gradients (dy/dx)

x.grad # Saves the gradient
````

**Stopping Gradient Tracking**

Sometimes, you dont want pytorch to track gradients, lets say during inference. You can decide to disable autograd.

````python
x  = torch.tensor(3.0, requires_grad = True)
with torch.no_grad(): # This will disable the autograd
  y = x**2
  print(y.requires_grad)
````

**Detaching a tensor from computing Graph**

If you want to remove a tensor from tracking ggraidients, use `.detach()`

````python
x = torch.tensor(5.0, requires_grad = True)
y = x**2

y_detached = y.detach()
y_detached.requires_grad
````
